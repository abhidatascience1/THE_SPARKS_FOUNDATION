# Standardizing data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data=sc.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```


# To find the optimum number of clusters, we have used ELBOW METHOD

```python
wcss=[]
for k in range(1,11):
    km=KMeans(n_clusters=k,random_state=1)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss,color="r")
plt.title("Elbow Method")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS")
plt.grid()
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer as ev
elbow=ev(km,k=range(1,11))
elbow.fit(df)
elbow.poof()
plt.show()
```

```python
# Finding the optimum number of clusters for k-means classification
x = df.iloc[:, [0, 1, 2, 3]].values
from sklearn.cluster import KMeans
sse = []
for i in range(1,11):
    km = KMeans(n_clusters = i , random_state = 0)
    km.fit(x)
    sse.append(km.inertia_)
plt.plot(range(1,11), sse, color = 'red' , marker = '*')
plt.title("value of K")
plt.xlabel("k")
plt.ylabel("sum of square error")
plt.show()
```

#### The optimum number of cluster is the one where the graph bends and WCSS(within-cluster sun of square) doesn't decrease significantly after that point. Here  optimum value of k is 3

```python
# fitting the K-Means with using the value of k = 3
model = KMeans(n_clusters = 3 , random_state = 0)
y_means = model.fit_predict(x)
y_means
```

```python
plt.scatter(x[y_means == 0,0],x[y_means == 0,1],c = 'red', s= 50 , label = 'Iris-setosa')
plt.scatter(x[y_means == 1,0], x[y_means == 1 , 1] , c = 'blue' ,s=50, label = 'Iris-versicolour')
plt.scatter(x[y_means == 2,0] , x[y_means == 2,1] , c = 'green', s = 50 , label = 'Iris-virginica')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],marker = 's' , c = 'black' , label = 'Centriods')
plt.title("prediction of iris dataset")
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')


plt.legend()
```
